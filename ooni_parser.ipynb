```python
import fsspec
import requests
import json
import pandas as pd

### Edit your search parameters

limit = 1000 #<- how many results you want to return
offset = '0' # <---- Where you want the results to start 
test_type = 'web_connectivity'
from_date = '2022-11-01'
to_date = '2022-12-15'
site = requests.get("https://api.ooni.io/api/v1/measurements?limit="+str(limit)+"&failure=false&test_name="
+test_type+"&since="+from_date+"&until="+to_date+"&offset=" + str(offset))


## This is where the magic is happening
a = site.text
c = json.loads(a)
b = json.dumps(c['results'])
a = json.loads(b)
if test_type is 'web_connectivity':
  dict1 = {}
  dict2 = {}
  ctr = 0
  for i in a:
    if 'blocking' in str(i['scores']):
      for scores in i['scores']:
        if 'type' in str(i['scores'][scores]):
          for analysis in i['scores'][scores]:
            dict2[analysis] = i['scores'][scores][analysis]
        else:
          dict2[scores] = i['scores'][scores]
    dict1[ctr] = i
    dict1[ctr].update(dict2)
    ctr+=1
list_of_dicts = []    
for i in dict1:
  list_of_dicts.append(dict1[i])

df = pd.DataFrame(list_of_dicts)


### Change the name of file to whatever you want

# Add this if you want only results without null parameters
#df = df.dropna()



df.to_csv('ooni_test.csv')


    
print(df.head)    
    



```
